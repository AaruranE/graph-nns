```python
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
records_filename = "../src/records.csv"
p = 0.05
df = pd.read_csv(records_filename,header=0, 
                 skiprows=lambda i: i>0 and random.random() > p)
```

```python
plt.hist(df['test_acc'])
plt.grid()
plt.title("Test accuracy - Histogram")
plt.show()
```

```python
plt.hist(df['test_loss'])
plt.grid()
plt.title("Test Loss - Histogram")
plt.show()
```

```python
plt.scatter(df['test_loss'], df['test_acc'])
plt.grid()
plt.
plt.title('Test Accuracy vs Test Loss')
plt.show()
```
